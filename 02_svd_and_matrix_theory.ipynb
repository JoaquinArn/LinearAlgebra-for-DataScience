### Conceptos clave previos

$\underline{Teorema}: \text{(Descomposición de Shur)}$

$A \in \mathbb{K}^{n \times n} \quad \lambda_1, ..., \lambda_n$ autovalores (contados con multiplicidad)

$\Rightarrow \exists \ U$ unitaria y $T$ triangular superior tal que $A = UTU^*$

$\hspace{3cm} $ Además: $T_{ii} = \lambda_i$

$\underline{Def}: \ A \in \mathbb{C}^{n \times n}$ se dice unitariamente equivalente a $B$ si $A = UBU^*$ para alguna $U$ unitaria.

\

$\underline{Def}: \  A \in \mathbb{K}^{n \times n}$ se dice normal si $AA^* = A^*A$

$Obs: A $ unitaria y/o hermitiana (o simétrica si es una matriz real) $\rightarrow A$ normal.

\

$\underline{Lema}: \ A$ normal, $B$ unitariamente equivalente a $A \rightarrow B$ normal. \
$D/ \quad B^*B = (UAU^*)^*UAU^* = UA^*\underbrace{U^*U}_IAU = UA^*AU \underbrace{=}_{A \text{ normal}} UAA^*U = UAU^*(UAU^*)^* = BB^* $

$\Rightarrow B$ normal.

$\hspace{18cm} \boxtimes$

\

$\underline{Teorema}: \ A \in \mathbb{C}^{n \times n}$ normal $\rightarrow A = UDU^*$ con $U$ unitaria y $D$ matriz diagonal cuyos elementos son los autovalores de $A$ (con su multiplicidad).

$D/ \quad$ Planteo $A = UTU^*$ su descomposición de Shur

Luego, eso significa que $U^*AU = T$

Lo cual, al ser $A$ normal, implica que, por el lema enunciado, $T$ también lo es.

$\Rightarrow T^*T = TT^*$

Desarrollo (teniendo en cuenta $T = \begin{bmatrix} t_{11} & t_{12} & . & . & . & t_{1n} \\
0 & t_{22} & t_{23} & . & . & t_{2n} \\
. & 0 & . & & & .\\. &.&& . && . \\ .&.&&& . & . \\ 0 & 0 & . & . & 0 & t_{nn}\end{bmatrix}$)

\

$\begin{bmatrix} \bar{t_{11}} & 0 & . & . & . & 0 \\
\bar{t_{12}} & \bar{t_{22}} & 0 & . & . & 0 \\
. & & . & & & .\\. &&& . && . \\ .&&&& . & 0 \\ \bar{t_{1n}} & \bar{t_{2n}} & . & . & . & \bar{t_{nn}}\end{bmatrix}$
$\begin{bmatrix} t_{11} & t_{12} & . & . & . & t_{1n} \\
0 & t_{22} & t_{23} & . & . & t_{2n} \\
. & 0 & . & & & .\\. &.&& . && . \\ .&.&&& . & . \\ 0 & 0 & . & . & 0 & t_{nn}\end{bmatrix}$ = $\begin{bmatrix} t_{11} & t_{12} & . & . & . & t_{1n} \\
0 & t_{22} & t_{23} & . & . & t_{2n} \\
. & 0 & . & & & .\\. &.&& . && . \\ .&.&&& . & . \\ 0 & 0 & . & . & 0 & t_{nn}\end{bmatrix}$ $\begin{bmatrix} \bar{t_{11}} & 0 & . & . & . & 0 \\
\bar{t_{12}} & \bar{t_{22}} & 0 & . & . & 0 \\
. & & . & & & .\\. &&& . && . \\ .&&&& . & 0 \\ \bar{t_{1n}} & \bar{t_{2n}} & . & . & . & \bar{t_{nn}}\end{bmatrix}$

\

Fila 1 Columna 1: $\require{enclose} \enclose{updiagonalstrike}{\bar{t_{11}} * t_{11}} = \enclose{updiagonalstrike}{t_{11}*\bar{t_{11}}} + t_{12}*\bar{t_{12}} + ... + t_{1n}*\bar{t_{1n}}$

$\hspace{5.05cm} 0 = |t_{12}|^2 + ... + |t_{1n}|^2$

Lo cual puede ocurrir solamente si $ \ t_{12} = ... = t_{1n} = 0$.

Fila 2 Columna 2: $\quad \underbrace{\bar{t_{12}}*t_{12}}_{t_{12} = 0} + \enclose{updiagonalstrike}{\bar{t_{22}} * t_{22}} = \enclose{updiagonalstrike}{t_{22}*\bar{t_{22}}} + t_{23}*\bar{t_{23}} + ... + t_{2n}*\bar{t_{2n}}$

$\hspace{7.88cm} 0 = |t_{23}|^2 + ... + |t_{2n}|^2$

Lo cual puede ocurrir solamente si $ \ t_{23} = ... = t_{2n} = 0$.

\

$\rightarrow $ Repitiendo la cuenta para todas las filas, pruebo que $T$ es diagonal.

$\hspace{18cm} \boxtimes$

$\underline{Corolario}: \ $ 1. $A$ normal $\rightarrow A$ diagonalizable y además existe $b.o.n.$ de autovectores \
$\hspace{2.75cm}$ 2. $A$ hermitiana (o $A \in \mathbb{R}^{n \times n}$ simétrica) $\rightarrow A$ diagonalizable con autovalores reales y $b.o.n$ de autovectores (reales en el caso de $A$ real simétrica).

\

Supongamos tenemos $A \in \mathbb{R}^{n \times n}$ simétrica, $\{ u_1,...,u_n \}\ \text{b.o.n.}$ de autovectores de $A$.

$\hspace{9cm} \lambda_1,...,\lambda_n $ sus correspondientes autovalores

Tomo $v \in \mathbb{R}^n$ cualquiera, $v = \alpha_1u_1 + ... + \alpha_nu_n$

$\hspace{3cm} ||v||_2^2 = v^tv = (\alpha_1u_1 + ... + \alpha_nu_n)^t \ (\alpha_1u_1 + ... + \alpha_nu_n) = (\alpha_1u_1^t + ... + \alpha_nu_n^t)(\alpha_1u_1 + ... + \alpha_nu_n)$

$\hspace{4.15cm} = \sum \limits_{i=1}^n \ \sum \limits_{j=1}^n \alpha_i \ \alpha_j \underbrace{(u_i^tu_j)}_{\left\{
\begin{array}{ll}
0 & \text{si } i \neq j \\
||u_i||_2^2 = 1 & \text{si } i = j
\end{array}
\right.} = \sum \limits_{i=1}^n \alpha_i^2 \quad \text{Teorema de Pitágoras}$

\

$Av = A(\alpha_1u_1 + ... + \alpha_nu_n) = (\alpha_1 \lambda_1)u_1 + ... + (\alpha_n \lambda_n)u_n$

$\frac{||Av||_2^2}{||v||_2^2} = \frac{\sum \limits_{i=1}^{n} (\alpha_i \lambda_i)^2}{\sum \limits_{i=1}^n \alpha_i^2} \leq \frac{\sum \limits_{i=1}^{n} \alpha_i^2 \lambda_{max}^2}{\sum \limits_{i=1}^n \alpha_i^2} \quad $ donde $\lambda_{max} = \{ \text{autovalor} \ / \ |\lambda_{max}| \geq |\lambda_i| \ \forall i \}$

$\frac{\sum \limits_{i=1}^{n} \alpha_i^2 \lambda_{max}^2}{\sum \limits_{i=1}^n \alpha_i^2} = \lambda_{max}^2 \ \enclose{updiagonalstrike}{\frac{\sum \limits_{i=1}^n \alpha_i^2}{\sum \limits_{i=1}^n \alpha_i^2}} = \lambda_{max}^2$

$\rightarrow \frac{||Av||_2}{||v||_2} \leq |\lambda_{max}| \quad$ o sea que aún con el $v$ tal que logre maximizar la expresión, el resultado no podrá superar el módulo del autovalor de mayor módulo de $A$.

Luego, si tomamos $\lambda_{max} = \lambda_k$:

$\frac{||Au_k||_2}{||u_k||_2} = \frac{||\lambda_k u_k||_2}{1} = |\lambda_k| \ \underbrace{||u_k||_2}_{=1} = |\lambda_k| = |\lambda_{max}|$

Por lo tanto: $\quad \underset{v \neq 0}{max} \frac{||Av||_2}{||v||_2} = |\lambda_{max}| \ \rightarrow ||A||_2 = |\lambda_{max}|$

\



Si $A$ no es simétrica, al realizar $A^tA$, la cual es simétrica pues $(A^tA)^t = (A^t)^tA^t = A^tA$, entonces $||A^tA||_2 = |\lambda_{max_{(A^tA)}}|$ \
Luego, como $||A||_2 = ||A^t||_2$ entonces $|\lambda_{max_{(A^tA)}}| = ||A^tA||_2 \leq ||A||_2^2 \ $ y, por lo tanto, $||A||_2 \geq \sqrt{|\lambda_{max_{(A^tA)}}|}$.


Por otro lado, sea $x \in \mathbb{R}^n$, y sea $\{ v_1, ..., v_n \} \ \text{b.o.n.}$ de autovectores de $A^tA$ asociados a los correspondientes autovectores $\lambda_{1_{(A^tA)}}, ..., \lambda_{n_{(A^tA)}}$; con:

$\hspace{6cm} A^tAv_i = \lambda_{i_{(A^tA)}}v_i$

Entonces:

$\hspace{3cm} x = \sum \limits_{i=1}^n \alpha_i v_i$

Al evaluar $x^tA^tAx$:

$x^tA^tAx = (\sum \limits_i \alpha_i v_i)^tA^tA(\sum \limits_j \alpha_j v_j) = \sum \limits_{i,j} \alpha_i \alpha_j v_i^tA^tAv_j$

Como $v_i \perp v_j$ si $i \neq j$;$\ \ $ y $\ A^tAv_j = \lambda_{j_{(A^tA)}} v_j$, entonces:

$\hspace{7cm} v_i^tA^tAv_j = \lambda_{j_{(A^tA)}} v_i^t v_j = {\left\{
\begin{array}{ll}
0 & \text{si } i \neq j \\
\lambda_{i_{(A^tA)}} \underbrace{||v_i||_2^2}_1 & \text{si } i = j
\end{array}
\right.} $

Lo que implica que: \
$x^tA^tAx = \sum \limits_i \lambda_i \alpha_i^2 \leq \sum \limits_i |\lambda_{max_{(A^tA)}}|\alpha_i^2 = |\lambda_{max_{(A^tA)}}| \sum \limits_i \alpha_i^2 = |\lambda_{max_{(A^tA)}}|||x||_2^2$

Es decir que $x^tA^tAx = (Ax)^t(Ax) = ||Ax||_2^2 \leq |\lambda_{max_{(A^tA)}}| ||x||_2^2$

Y, como la norma 2 de $A$ se define por: \
$||A||_2 = \underset{||x||_2 = 1}{sup} ||Ax||_2$

Luego $||A||_2^2 \leq \lambda_{max_{(A^tA)}} \ \rightarrow ||A||_2 \leq \sqrt{|\lambda_{max_{(A^tA)}}|}$.

Por lo tanto, si $||A||_2 \geq \sqrt{|\lambda_{max_{(A^tA)}}|}\ $ y $ \ ||A||_2 \leq  \sqrt{|\lambda_{max_{(A^tA)}}|} \ \rightarrow ||A||_2 = \sqrt{|\lambda_{max_{(A^tA)}}|}$


\

$\underline{Def}:$ el radio espectral de una matriz $A$ es $\rho(A) = máx \{ |\lambda| \ / \ \lambda \text{ autovalor de A}\}$

Entonces, como he probado:

$\underline{Teorema}: A \in \mathbb{R}^{n \times n}$ \
- $A$ simétrica $\rightarrow ||A||_2 = \rho(A)$
- $A$ cualquiera $\rightarrow ||A||_2 = \sqrt{\rho(A^tA)}$

\

$\underline{Prop}: $ sea $A \in \mathbb{R}^{n \times n}$ (estrictamente) definida positiva, entonces todos los autovalores $\lambda$ de $A$ cumplen $\lambda \ (>) \geq 0$

\

$Obs: A \in \mathbb{R}^{n \times n} $ simétrica y definida positiva, entonces todos los autovalores $\lambda$ de $A$ son reales (pues $A \in \mathbb{R}^{n \times n}$ simétrica implica que $A$ es normal y, por lo tanto, diagonalizable con autovalores reales) no negativos (dado que la matriz es definida positiva).


## Descomposición en valores singulares (**SVD**)

La descomposición en valores singulares de una matriz $A \in \mathbb{C}^{m \times n}$ es un producto de la forma \
$A = U\Sigma V^* = \overset{\text{caso $m \ \geq \ n$}}{\overbrace{\underbrace{\begin{bmatrix} |& & & & & & & & |\\  |& & & & & & & & | \\ |&&& &&&& & |\\ u_1&.&.& .& .&.&.& .&u_m \\ |&&&&&& &  &| \\ |&&&& & && & |\\ |&&&& && & & |\\ |&  &  &  & & && & | \end{bmatrix}}_{U} \underbrace{\begin{bmatrix}
\sigma_1 & 0        & \cdots & 0 \\
0        & \sigma_2 & \cdots & 0 \\
\vdots   & \vdots   & \ddots & \vdots \\
0        & 0        & \cdots & \sigma_n \\
0        & 0        & \cdots & 0 \\
\vdots   & \vdots   &        & \vdots \\
0        & 0        & \cdots & 0
\end{bmatrix}}_{\Sigma} \underbrace{\begin{bmatrix}\ - & - & v_1^*& -&- \\ - & - & v_2^*& -&- \\ & &.&& \\ & & .&&  \\ &&.&&\\ -&-& v_n^*&-&- \end{bmatrix}}_{V^*}}} \ \ \ \lor \ \ \ \overset{\text{caso $n \ > \ m$}}{\overbrace{\underbrace{\begin{bmatrix}    |& & & &|  \\ |&&&&| \\ u_1&.&.& .&u_m \\ |&&&& | \\| &&& &|  \end{bmatrix}}_{U} \underbrace{\begin{bmatrix}
\sigma_1 & 0        & \cdots & 0        & 0        & \cdots & 0 \\
0        & \sigma_2 & \cdots & 0        & 0        & \cdots & 0 \\
\vdots   & \vdots   & \ddots & \vdots   & \vdots   &        & \vdots \\
0        & 0        & \cdots & \sigma_m & 0        & \cdots & 0
\end{bmatrix}}_{\Sigma} \underbrace{\begin{bmatrix}\ - & -& - & v_1^*& -&-&- \\-& - & - & v_2^*& -&- &-\\& & &.&& &\\& & & .&& & \\ &&&.&&&\\ -&-&-& v_n^*&-&-&- \end{bmatrix}}_{V^*}}} $ \
donde $U \in \mathbb{C}^{m \times m}, V \in \mathbb{C}^{n \times n}$ son unitarias y $\Sigma \in \mathbb{C}^{m \times n} $ es una matriz diagonal con todas sus entradas reales y no negativas. Más aún:

- Las entradas de la diagonal de $\Sigma$ son $\sigma_1,...,\sigma_n$ (con $\sigma_1 \geq... \geq \sigma_n$) donde $\sigma_1^2,...,\sigma_n^2$ son los autovalores de $A^*A$ la cual es:
  - Hermitiana: $(A^*A)^* = A^* (A^*)^* = A^*A$
  - Definida positiva (no estricta): $v \in \mathbb{C}^n \rightarrow v^*A^*Av =(Av)^*(Av) = ||Av||_2^2 \geq 0$

  Esto implica que $\sigma_1, ..., \sigma_n \geq 0$.

- Las columnas de $V$ son los autovectores correspondientes $v_1,...,v_n$ (es decir, los autovectores de $A^*A$ asociados a $\sigma_1^2,...,\sigma_n^2$).

- Las columnas de $U: u_1,...,u_n,...,u_m$; vienen dadas por la relación $Av_j = \sigma_j u_j \ \ \forall j = 1,...,n$
  - En este sentido, si $\sigma_j = 0$ para algún $j \leq min \{m,n \}$ entonces se puede elegir $u_j$ de modo que se preserve la ortonormalidad de las columnas de $U$.

\

$\underline{Definición}:$ dada una matriz $A \in \mathbb{C}^{m \times n}$ y su descomposición $ \ S. \ V.D.$ , definimos su pseudoinversa como: $A^{\dagger} = V\Sigma^{\dagger}U^*$

$ = \underbrace{\begin{bmatrix}| &  &  &&& | \\ | &&&&& | \\ | &&&&& | \\ v_1 & . & .& . & .& v_n \\ | &&&&& | \\ | &&&&& | \\ | &&&&& | \end{bmatrix}}_{n \times n} \underbrace{\begin{bmatrix}\frac{1}{\sigma_1} &  &  & &&& &| & 0 ... 0 \\  & . &&&&&&| & .\\ && \frac{1}{\sigma_s}& &&&& | & .\\ &&& 0 &&&&| & .\\ &&&&.&& &|  & .\\ &&&&& . &&| & .\\ &&&&&& 0 & | & 0...0 \end{bmatrix}}_{n \times m} \underbrace{\begin{bmatrix}\ -& - & - & - & u_1^*& -&- &- & - \\  -& - & - & - & u_2^*& -&- &- & - \\ &&& &.&&& & \\ &&& & .&&& & \\ &&&&.&& &  & \\ &&&& .& && & \\ &&&& .&& & & \\  -& - & - & - & u_m^*& -&- &- & - \end{bmatrix}}_{m \times m}$

donde $\sigma_s$ es el último valor positivo y, por ende, $\sigma_{s+1} = ...= \sigma_n = 0$

> **Nota**: La matriz $\Sigma^{\dagger}$ representada anteriormente fue construida bajo la suposición $m \geq n$ (caso habitual). En este contexto, los valores singulares $\frac{1}{\sigma_i}$ están contenidos en una submatriz diagonal de dimensión $n \times n$, y el bloque restante de ceros se ubica en las **últimas columnas**, completando el formato rectangular de tamaño $m \times n$.

> En el caso alternativo donde $n > m$, la estructura de $\Sigma^{\dagger} \in \mathbb{C}^{m \times n}$ se adapta colocando los ceros **como filas adicionales** por debajo del bloque diagonal. Es decir, los valores singulares se distribuyen sobre una diagonal de dimensión $m \times m$, y el resto de la matriz se completa con ceros en las **filas inferiores**.

Esta matriz cumple ciertas propiedades:
- $A A^{\dagger}A = A$
- $A^{\dagger} A A^{\dagger} = A^{\dagger}$
- $(AA^{\dagger})^* = AA^{\dagger}$
- $(A^{\dagger}A)^* = A^{\dagger}A$
- $A^{\dagger}$ es única.

\

$Obs: A \in \mathbb{K}^{n \times n}$ es inversible $\iff \sigma_i \neq 0 \ \  \forall i$ \
Esto se puede probar de varias maneras. Lo propongo desde el enfoque del determinante: \
$A$ inversible $\rightarrow A^*$ inversible \
Eso implica que $det(A), \ det(A^*) \neq 0$ \
$\Rightarrow det(A^*A) = \underbrace{det(A^*)}_{\neq 0}\underbrace{det(A)}_{\neq 0} \neq 0 \rightarrow A^*A$ inversible. \
Por lo tanto, como $A^*A$ es inversible, entonces se cumple que para todo autovalor $\lambda$ de $A^*A$, $ \ \lambda \neq 0$. Luego, como $\sigma_1^2, ..., \sigma_n^2$ son los autovalores de $A^*A$ (o sea $\sigma_i = \sqrt{\lambda_i}$, donde $\lambda_i > 0$ pues $A$ simétrica, definida positiva e inversible) $\Rightarrow \sigma_1,...\sigma_n \neq 0$. \
Esto lleva a que: \
$Obs:$ sea $A \in \mathbb{K}^{n\times n}$ inversible, $A = U \Sigma V^*$ entonces $A^{-1} = V \Sigma^{-1}U^* = V \Sigma^{\dagger}U^* = A^{\dagger}$ \
Computacionalmente, esto significa que si se cuenta con la descomposición en valores singulares de una matriz inversible (a priori quizá no se conoce si lo es, pero se puede verificar fácilmente observando el valor singular más chico $\sigma_n$ de la matriz $A$, o sea, el último elemento de la diagonal de $\Sigma$), entonces el costo de invertirla es simplemente dar vuelta (invertir) cada elemento de la diagonal de $\Sigma$ (es decir, $\Theta(n)$).

\

$\underline{Teorema}: \ A \in \mathbb{K}^{m \times n} \rightarrow ||A||_2 = \sigma_1$ \

$D / \quad ||A||_2 = \sqrt{\rho (A^tA)} = max \{ \sqrt{\lambda} : \lambda \text{ autovalor de } A^tA \} = max \{\sigma : \sigma \text{ valor singular de } A \} = \sigma_1$ \
$\hspace{21cm} \boxtimes$

\

$Obs: \ A \in \mathbb{R}^{n \times n}$ inversible, $\sigma_1 \geq...\geq \sigma_n$ valores singulares de $A \rightarrow cond_2(A) = ||A||_2||A^{-1}||_2 = \frac{\sigma_1}{\sigma_n}$

Esto permite generalizar la noción de $cond_2(A)$ a matrices del estilo $A \in \mathbb{K}^{m \times n}$ de rango completo $s = min \{ m,n \}$. \
Se define entonces $cond_2(A) = \frac{\sigma_1}{\sigma_s}$.

\

Adicionalmente se puede definir la $\ S. \ V.D.$ truncada de rango $k \ (k \leq min \{m,n \})$ quedándose hasta el $k$-ésimo valor singular en la diagonal, y haciendo que los demás sean $0$. Si llamamos $A_k$ a esa matriz, entonces:

$\hspace{5cm} A_k = \sum \limits_{i=1}^k \sigma_i u_i v_i^*$

\

$\underline{Proposición}:$ la descomposición $S. \ V.D.$ truncada en $k \leq min\{m,n\}$ es la mejor aproximación de rango $k$ que se puede hacer de $\ A$, es decir:

$A_k = arg \ min \{||A-B||_2 : rk(B) = k \}$

Donde $arg \ min \{||A-B||_2 : rk(B) = k \} = \sigma_{k+1}$

**Ejercicio SVD** \
Escribir un programa que reciba como input una matriz $A \in \mathbb{R}^{m \times n}$ y un entero positivo $r$ y:
 - Calcule la descomposición en valores singulares de $A$.
 - Devuelva: una tupla con la dimensión original de la matriz $A (n,m)$, las matrices $\tilde{U} \in \mathbb{R}^{m× r}$ y  $\tilde{V} \in \mathbb{R}^{r \times n}$ que surgen de eliminar de $U$ y $V$ los vectores singulares con índice mayor a $r$ y el vector de valores singulares $\tilde{s} \in \mathbb{R}^{r}$, también recortado.

```python
# Debo crear la función que devuelva, a partir de A de dimensiones m x n, A_k con k <= min(m,n)

def svd_truncada (A, r):
  #Primero calculo la SVD de A con el comando de numpy np.linalg.svd
  #Este devuelve la matriz U de m x m, V de n x n y s de dimensión min{m,n} tal que A = Udiag(s)V
  U, s, V_t = np.linalg.svd(A)

  #Luego, me quedo con los primeros r elementos de s; las primeras r columnas de U; y las primeras r filas de V
  s_tilde = s[:r]
  U_tilde = U[:,:r]
  V_tilde_t = V_t[0:r,:]

  #Finalmente, devuelvo la dimensión original de A; y U_tilde, s_tilde, y V_tilde (o sea, V_tilde_t traspuesta)

  return A.shape, U_tilde, s_tilde, V_tilde_t.T
```

**Ejercicio**\
Escribir un programa que recontruya la matriz a partir del output del programa anterior. Es decir, que:
 - Reciba el tamaño $(m,n)$, las matrices $\tilde{U}$ y $\tilde{V}$, y el vector $\tilde{s}$.
 - Amplie las matrices con ceros generando $U' \in \mathbb{R}^{m \times m}, V' \in \mathbb{R}^{n \times n}$; y ponga $\tilde{s}$ en la diagonal de una matriz $Σ' \in \mathbb{R}^{m \times n}$.
 - Devuelva la matriz $B = U'Σ'V' $ (que es la mejor aproximación a $A$ entre las matrices de rango $r$).

```python
# Debo programar una función que devuelva la mejor aproximación de rango r de la matriz A a partir de su SVD truncada

def mejor_aproximacion_menor_rango (tamanio, U_monio, V_monio, s_monio):
  #Primero desgloso el tamaño pasado como parámetro
  m = tamanio[0]
  n = tamanio[1]
  #Formo U_tilde, V_tilde y sigma_tilde; en principio, con ceros.
  U_tilde = np.zeros( (m,m))
  V_tilde = np.zeros((n,n))
  sigma_tilde = np.zeros((m,n))
  #Ahora, recorro las matrices como parámetro, y completo las matrices _tilde con la información de las variables pasadas como parámetro.
  for i in range(len(s_monio)):
    U_tilde[:, i] = U_monio[:, i]
    V_tilde[:, i] = V_monio[:, i]
    sigma_tilde[i, i] = s_monio[i]

  #Finalmente, multiplico las matrices y devuelvo la matriz resultante
  return U_tilde @ sigma_tilde @ V_tilde.T
```
